## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df.boxplot(column =['ConvertedComp'],grid=False)  
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
data = np.array(df['ConvertedComp'])
q3, q1 = np.percentile(data, [75, 25])
iqr = q3 - q1
print('iqr is ',iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
upperbound = q3 + (1.5*iqr)
lowerbound = q1 - (1.5*iqr)
print('upper bound is ', upperbound)
print('lower bound is ',lowerbound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
print((df['ConvertedComp'] > upperbound).sum())
print((df['ConvertedComp'] < lowerbound).sum())
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df1 =df[(df['ConvertedComp'] > (lowerbound)) & (df['ConvertedComp'] < (upperbound))]
```
